```python
#Importing Pandas
import pandas as pd
```

```python
#loding Netflix data set (CSV File)
df = pd.read_csv('netflix_titles.csv')
df
```

```python
# Printing first 5 records
df.head()
```

```python
# Printing last 5 records
df.tail()
```

```python
# Size of data (count of rows & columns)
col , row = df.shape
print("column count:", col)
print("Row count:", row)
```

```python
# Column information
df.info()
```

```python
# Check for missing values
df.isnull().sum()
```

```python
# Handling Missing Values

# Filling String -based missing values
df["director"] = df["director"].fillna("Unknown")
df["cast"] = df["cast"].fillna("Not Listed")
df["country"] = df["country"].fillna("Unknown")
df["rating"] = df["rating"].fillna("Unrated")
df["duration"] = df["duration"].fillna("Unknown")

# dropping missing rows from 'date_added' (only 10 rows)
df.dropna(subset="date_added",inplace=True)
```

```python
# After Handling Missing Values
df.isnull().sum()
```

```python
df.head()
```

```python
# Displaying Movies from "India"
df[(df['country']=='India') & (df['type'] == 'Movie')]
```

```python
# Displaying shows rated "TV-MA"
df[(df['type']=='TV Show') & (df['rating'] == 'TV-MA')]
```

```python
# Displaying titels releaseed after 2015
df[df['release_year'] > 2015]
```

```python
# sorting data by title and release_year
df.sort_values(by = ['title','release_year'], inplace=True )
df
```

```python
#  Group by type
df['type'].value_counts()
```

```python
# Group by Country
df['country'].value_counts()
```

```python
# bar plot for types
import matplotlib.pyplot as plt

df['type'].value_counts().plot(kind='bar', color=['#1f77b4', '#ff7f0e'])
plt.title('Count of Movies vs TV Shows')
plt.xlabel('Type')
plt.ylabel('Count')
plt.show()
```
